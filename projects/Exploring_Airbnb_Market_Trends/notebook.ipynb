Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import necessary packages
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Begin coding here ...
# Use as many cells as you like
prices_data = pd.read_csv("data/airbnb_price.csv")
desc_data = pd.read_excel("data/airbnb_room_type.xlsx")
reviews_data = pd.read_csv("data/airbnb_last_review.tsv", sep="\t")
```

```python
print(reviews_data.info())
```

```python
print(reviews_data.head())
```

```python
reviews_data["last_review"] = pd.to_datetime(reviews_data["last_review"]).dt.date
print(reviews_data["last_review"].dtype)
print(reviews_data.head())
```

```python
print(reviews_data.sort_values("last_review", ascending=False))
```

```python
early_reviews = reviews_data[
    reviews_data["last_review"] <= dt.date(2019, 4, 30)
][["listing_id", "last_review"]]
print(early_reviews.head())
```

```python
early_reviews = reviews_data["last_review"].min()
```

```python
recent_reviews = reviews_data[
    reviews_data["last_review"] > dt.date(2019, 4, 30)
][["listing_id", "last_review"]]
print(recent_reviews.head())
```

```python
recent_reviews = reviews_data["last_review"].max()
```

```python
print(desc_data.head())
```

```python
print(desc_data["room_type"].value_counts())
desc_data["room_type"] = desc_data["room_type"].str.strip()
matchings = {
    "Private room" : "private room",
    "PRIVATE ROOM" : "private room" 
}
desc_data["room_type"] = desc_data["room_type"].replace(matchings)
print(desc_data["room_type"].value_counts())
```

```python
privates_airbnb = desc_data[desc_data["room_type"] == "private room"][["listing_id", "room_type"]]
print(privates_airbnb.head())
```

```python
print(prices_data.head())
print(prices_data.info())
```

```python
prices_data["price"] = prices_data["price"].str.strip(" dollars")
prices_data["price"] = prices_data["price"].astype("int")
print(prices_data.info())
```

```python
avg_price = round(prices_data["price"].mean(), 2)
print(avg_price)
```

```python
review_dates = {
    "first_reviewed" : [early_reviews],
    "last_reviewed" : [recent_reviews],
    "nb_private_rooms" : [len(privates_airbnb)],
    "avg_price" : [avg_price]
}

print(review_dates)
review_dates = pd.DataFrame(review_dates)
print(review_dates)
```
